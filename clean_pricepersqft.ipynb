```python
import pandas as pd
import numpy as np
```

```python
pricepersq = pd.read_csv('./db/pricepersqft.csv')
pricepersq.head()
```

```python
# statck the columns of year and month into one column and add a column to indicate month-year
pricepersq.set_index(['City Code','City','Lat','Lon','Metro','County','State','Population Rank'],inplace = True)
pricenew = pricepersq.stack(dropna = False).reset_index()
```

```python
# separate month-year to two columns
pricenew.rename(columns={'level_8':'month_year',0:'price_persq'},inplace = True)
map_month_to_num = {'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'May':5,'Jun':6,'Jul':7,'Aug':8,'Sep':9,'Oct':10,'Nov':11,'Dec':12}
```

```python
pricenew['month'] = pricenew['month_year'].apply(lambda x: x[:3]).map(map_month_to_num)
```

```python
pricenew['year'] = pricenew['month_year'].apply(lambda x: '20' + x[4:]).astype(int)
```

```python
#pricenew.replace(np.nan,0,inplace = True)
```

```python
# filter out the year before 2012
pricenew_filter = pricenew[pricenew['year']>=2012]
```

```python
#pricenew_filter.fillna(0,inplace = True)

#pricenew.to_csv('pricepersqft_clean.csv')
```

```python
df=pricenew_filter
```

```python
df.sort_values(by=["City","year"]).head()
```

```python
ave_yearly_price=pricenew_filter.groupby(["City","year"])["price_persq"].mean().reset_index().rename(columns={"price_persq":"ave_price_persq"}).sort_values(by="ave_price_persq")
ave_yearly_price.reset_index(drop=True).head()
```
